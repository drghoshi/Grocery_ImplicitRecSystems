# Clusters for Instacart Dataset

```python
kmeans = KMeans(n_clusters=10000)
kmeans.fit(np_instacart_product_vectors)
len(kmeans.labels_)
labels_to_productid_dict = dict(zip(range(len(kmeans.labels_)), kmeans.labels_))
```

```python
with open('instacart_productid_to_clusterlabel_dict.pkl', 'wb') as f:
    pickle.dump(labels_to_productid_dict, f)
```
